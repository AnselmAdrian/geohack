```python
def rename_mnemonics(project):
    for well in pr:
        cols = []
        for i in range(len(well.header)):
            if well.header.section[i] == 'Curves':
                cols.append(well.header.descr[i].split(' ', 2)[1])
                # well.header.mnemonic[i] = well.header.descr[i].split(' ', 2)[1]
        well.data = dict(zip(cols[1:], list(well.data.values())))

rename_mnemonics(pr)
well: welly.well.Well = pr[0]
```

```python
def get_common_tracks(project):
    wells = {}
    for well in pr:
        key = next(filter(lambda x: x[0] == 'WELL', well.header.values))[3]
        track_names = {well.header.descr[i].split(' ', 2)[1] for i in range(len(well.header)) if well.header.section[i] == 'Curves'}
        x_loc = well.data['x_loc'].mean()[0]
        y_loc = well.data['y_loc'].mean()[0]
        wells[key] = {
                'track_names': track_names,
                'x_loc': x_loc,
                'y_loc': y_loc,
                }
    return wells
t = get_common_tracks(pr)
```

```python
all_track_names = set.union(*(x['track_names'] for x in t.values()))
common_track_names = set.intersection(*(x['track_names'] for x in t.values()))
contains_DTS = {key for key, value in t.items() if 'DTS' in value['track_names']}
```

```python
from collections import defaultdict
def track_name_count(tracks):
    counts = defaultdict(lambda: 0)
    for track in tracks.values():
        for track_name in track['track_names']:
            counts[track_name] = counts[track_name] + 1
    return counts
track_name_count(t)
        
```
